```python
from sympy import *
from matplotlib import pyplot as plt
```

```python
x = Symbol('x')
```

```python
f, g = Symbol('f(x)'), Symbol('g(x)')
```

```python
funcF = x**3 - 50*x
```

```python
funcG = -x**4 + 88 * x**2 - 241
```

```python
funcE = funcF - funcG
```

```python
funcE
```

```python
plot(funcF, funcG)
```

```python
plot(funcE)
```

```python
solve(funcE)
```

```python
q = solve(funcE)
```

```python
for i in range(len(q)):
    q[i] = q[i].evalf(5, chop = True)
```

```python
for i in range(len(q)):
    print(" : x =", q[i], "y = 0", sep=" ")
```
